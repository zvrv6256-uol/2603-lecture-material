# Part A — Equilibrium points and non-isolated equilibria

Consider the system:
dx/dt = 0  
dy/dt = -y

This system has infinitely many equilibrium points.


```python
xs = np.linspace(-2,2,20)
ys = np.linspace(-2,2,20)
X, Y = np.meshgrid(xs, ys)

U = np.zeros_like(X)
V = -Y

plt.figure(figsize=(5,5))
plt.quiver(X,Y,U,V)
plt.axhline(0,color='black')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Non-isolated equilibria")
plt.show()
```

# Part C — Limits of linearization

Consider the pendulum with friction (small angle approximation):
dθ/dt = ω  
dω/dt = -sin(θ) - 0.3ω


```python
def pendulum(t,z):
    theta, omega = z
    return [omega, -np.sin(theta) - 0.3*omega]

t = np.linspace(0,20,1000)
sol = solve_ivp(pendulum,(0,20),[1.5,0],t_eval=t)

plt.plot(sol.y[0], sol.y[1])
plt.xlabel("θ")
plt.ylabel("ω")
plt.title("Pendulum phase portrait")
plt.show()
```

# Part D — Lyapunov intuition

Consider candidate Lyapunov function:
V(x,y) = x^2 + y^2


```python
xs = np.linspace(-2,2,100)
ys = np.linspace(-2,2,100)
X,Y = np.meshgrid(xs,ys)
V = X**2 + Y**2

plt.contour(X,Y,V,levels=10)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Lyapunov function contours")
plt.show()
```
